#### 装饰器

```python
def run_time(func):
    def wrapper():
        start = time()
        func()                  # 函数在这里运行
        end = time()
        cost_time = end - start
        print("func three run time {}".format(cost_time))
    return wrapper
    
@run_time
def fun_one():
    sleep(1)
    
@run_time
def fun_two():
    sleep(1)
    
@run_time
def fun_three():
    sleep(1)
    
# 装饰器：函数的函数；具体使用是在函数前面加@装饰器的名字，表示对下面的函数进行装饰（增强）。
# 比如说：统计函数的运行时间。
```

```python
def quicksort(a,left,right):
    if left < right:
        base = location(a,left,right)
        quicksort(a,left,base-1)
        quicksort(a,base+1,right)
def location(a,left,right):
    base = a[left]
    while left < right:
        while left < right and a[right]>base:
            right -= 1
        if left < right:
            a[left] = a[right]
        while left < right and a[left]<base:
            left += 1
        if left < right:
            a[right] = a[left]
    a[left] = base
    return left
test = [5,2,3,1,4]
quicksort(test,0,len(test)-1)
print(test)
```

#### 归并排序 

```python
def mergesort(arr):
    # 边界条件
    if len(arr)<2:
        return arr
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    return merge(mergesort(left),mergesort(right))
def merge(left,right):
    result = []
    while len(left) and len(right):
        if left[0] <= right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    if len(left) > 0:
        result = result + left
    if len(right) > 0:
        result = result + right
    return result

a = [1,3,2,4,3]
print(mergesort(a))
```

#### 堆排序

```python
def swap(arr,a,b):
    c = arr[a]
    arr[a] = arr[b]
    arr[b] = c
    
def buildMaxHeap(arr):
    import math
    # 最后一个由子节点的节点
    for i in range(math.floor(len(arr)/2),-1,-1):
        heapify(arr,i)
        
def heapify(arr,i): # i一直能取到0
    left = 2*i+1
    right = 2*i+2
    largest = i
    if left < arrlen and arr[left]>arr[largest]:
        largest = left
    if right < arrlen and arr[right]>arr[largest]:
        largest = right
    if largest != i:
        swap(arr,i,largest)
        heapify(arr,largest) # 调整后往下走，看下面是否满足最大堆
    
def heapSort(arr):
    global arrlen
    arrlen = len(arr)
    buildMaxHeap(arr)
    for i in range(len(arr)-1,0,-1): #不包含第一个
        swap(arr,0,i)
        arrlen -= 1
        heapify(arr,0) # arrlen已经被用来筛选数组长度
    return arr
```

#### 应用1-第k大(小)元素

```python
### 遍历基本框架### 遍历# 易错点：第k大应该是索引为n-k处的元素
# -*- coding:utf-8 -*-

# -*- coding:utf-8 -*-

class Solution:
    def location(self,a,left,right):
        temp = left
        while left < right:
            while left < right and a[right] > a[temp]:
                right -= 1
            while left < right and a[left] < a[temp]:
                left += 1
            a[left],a[right] = a[right],a[left]
        a[temp],a[left] = a[left],a[temp]
        return left
    
    def quicksort(self,a,left,right,K):
        if left <= right:
            m = self.location(a,left,right)
            if m == K:
                return a[m]
            elif m > K:
                return self.quicksort(a, left, m-1, K)
            else:
                return self.quicksort(a, m+1, right, K)
        
    def findKth(self, a, n, K):
        # write code here
        # n-K处的数
        return self.quicksort(a, 0, n-1, n-K)
    
test = Solution()
a = [1,3,5,2,2]
n = 5 
K = 3
print(test.findKth(a,n,K))
```

#### 二分查找

```python
class Solution:
    def search(self , nums , target ):
        # write code here
        left = 0
        right = len(nums)-1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                while mid != 0 and nums[mid] == nums[mid-1]:
                    mid = mid - 1
                return mid
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1
        return -1
```

#### 遍历

> 跟回溯的遍历是一样的

```python
# 前序
def preorder(self,root):
    if not root:
        return 
    self.temp.append(root.val)
    self.preorder(root.left)
    self.preorder(root.right)

# 层次遍历
from collections import deque
class Solution:
    def levelOrder(self , root ):
        # write code here
        q = deque([])
        result = []
        if not root:
            return []
        q.append(root)
        while len(q) != 0:
            length = len(q)
            temp = []
            for i in range(length):
                node = q.popleft()
                temp.append(node.val)
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            result.append(temp)
        return result
```

```python
class Solution:
    def merge(self , A, m, B, n):
        # write code here
        k = m+n-1
        i = m-1
        j = n-1
        while i>=0 and j>=0:
            if A[i]>B[j]:
                A[k] = A[i]
                k -= 1
                i -= 1
            else:
                A[k] = B[j]
                k -= 1
                j -= 1
        while i>=0:
            A[k] = A[i]
            k -= 1
            i -= 1
        while j>=0:
            A[k] = B[j]
            k -= 1
            j -= 1
```

#### [数组中三数相加为0的元素](https://www.nowcoder.com/practice/345e2ed5f81d4017bbb8cc6055b0b711?tpId=188&tqId=38291&rp=1&ru=%2Factivity%2Foj&qru=%2Fta%2Fjob-code-high-week%2Fquestion-ranking&tab=answerKey)

```python
class Solution:
    def threeSum(self , num ):
        num.sort()
        result = []
        for i in range(len(num)):
            if num[i] > 0:
                break
            if i > 0 and num[i]==num[i-1]:
                continue
            l = i+1 if i < len(num)-1 else len(num)-1
            r = len(num)-1
            while l<r:
                if num[l]+num[r]+num[i] == 0:
                    result.append((num[i],num[l],num[r]))
                    l += 1
                    r -= 1
                    while l<r and num[l]==num[l-1] and num[r]==num[r+1]:
                        l += 1
                        r -= 1
                elif num[l]+num[r]+num[i] > 0:
                    r -= 1
                else:
                    l += 1
        return result
```

```python
# 注意result增加track时，要另外复制一个，因为指向的是同一个对象
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        def backtrack(nums,track):
            if len(track) == len((nums)):
                result.append(track[:])
                return
            for i in range(len(nums)):
                if nums[i] in track:
                    continue
                track.append(nums[i])
                backtrack(nums,track)
                track.pop()
        result = []
        backtrack(nums,[])
        return result
```

```python
# 全排列2 去除重复组合
class Solution:
    def permuteUnique(self, nums: List[int]) -> List[List[int]]:
        def backtrack(nums,track,select):
            if len(track) == len(nums):
                result.append(track[:])
                return
            for i in range(len(nums)):
                if (select[i]) or (i > 0 and nums[i] == nums[i-1] and not select[i-1]):
                    continue
                track.append(nums[i])
                select[i] = 1
                backtrack(nums,track,select)
                track.pop()
                select[i] = 0
        result = []
        nums.sort()
        select = [0]*len(nums)
        backtrack(nums,[],select)
        return result
```
