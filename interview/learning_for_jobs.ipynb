#### 快速排序

```python
# 不熟练，写的很慢
def location(arr,left,right):
    temp = left
    while left < right:
        while left < right and arr[right]>=arr[temp]:
                right -= 1
        while left < right and arr[left]<=arr[temp]:
                left += 1
        # 交换left和right
        c = arr[left] 
        arr[left] = arr[right]
        arr[right] = c
    # 把参考的数（arr[0]）放到left位置上
    c = arr[temp]
    arr[temp] = arr[left]
    arr[left] = c
    return left
    
def quicksort(arr,left,right):
    if left < right:
        temp = location(arr,left,right)
        # 剩余左右两边
        quicksort(arr,left,temp-1)
        quicksort(arr,temp+1,right)
```

#### 归并排序 

```python
def mergesort(arr):
    # 边界条件
    if len(arr)<2:
        return arr
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    return merge(mergesort(left),mergesort(right))
def merge(left,right):
    result = []
    while len(left) and len(right):
        if left[0] <= right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    if len(left) > 0:
        result = result + left
    if len(right) > 0:
        result = result + right
    return result

a = [1,3,2,4,3]
print(mergesort(a))
```

#### 堆排序

#### 应用1-第k大(小)元素

```python
# 易错点：第k大应该是索引为n-k处的元素
# -*- coding:utf-8 -*-

class Solution:
    def location(self,a,left,right):
        base = left
        while left < right:
            while left < right and a[right] > a[base]:
                right -= 1
            while left < right and a[left] < a[base]:
                left += 1
            temp = a[right]
            a[right] = a[left]
            a[left] = temp
        temp = a[base]
        a[base] = a[left]
        a[left] = temp
        return left
         
    def quicksort(self,a,left,right,K):
        temp = self.location(a, left, right)
        if temp == K:
            return a[temp]
        elif temp > K:
            return self.quicksort(a, left, temp-1, K)
        else:
            return self.quicksort(a, temp+1, right, K)

    def findKth(self, a, n, K):
        # write code here
        return self.quicksort(a, 0, n-1, n-K)
```

#### 层次遍历

```python
# 层次遍历
from collections import deque
class Solution:
    def levelOrder(self , root ):
        # write code here
        q = deque([])
        result = []
        if not root:
            return []
        q.append(root)
        while len(q) != 0:
            length = len(q)
            temp = []
            for i in range(length):
                node = q.popleft()
                temp.append(node.val)
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            result.append(temp)
        return result
```
