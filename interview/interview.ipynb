# 面试 

## 自我介绍 

### 第一种

基本信息其实简历上都有，面试官想知道的是你是否甚至这份工作。


### 第二种

3+1模式
- 能力+例子
- 性格：我是一个什么样的人
- 职业规划：想成为一个优秀的算法工程师。

## 介绍项目

问项目细节
怎么思考这个项目的，其中用了哪些方法；别人怎么做的；想过怎么改进？

## 最有成就感的事

拆解事情-目标-行动（具体怎么做，遇到什么问题，怎么解决）-最终达到什么结果

## 自己的优点

坚持做手账和记录10年，对于自己想做的事情，我会仔细研究，制作详细的计划，并一步一步实现。

## 数据分析要掌握的技能

学习机器学习，至少需要掌握一门编程语言+业务知识+一个大数据工具。必学python，c++作为第二语言。
