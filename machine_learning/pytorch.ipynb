## RNN

```python
import torch
```

一层RNN

```python
rnn=torch.nn.RNN(input_size=100,hidden_size=20,num_layers=1)
print(rnn)
x = torch.randn(10,3,100)
out,h = rnn(x,torch.zeros(1,3,20))
print(out.shape,h.shape)
```
